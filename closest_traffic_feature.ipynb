```python
%%sql

SELECT Juhtumi_nr, Name, SQRT (POW(Longitude - Lon, 2) + POW(Latitude - Lat, 2)) * 111.3 AS Kaugus FROM liiklusonnetus, seadmed LIMIT 150;
```

```python
%%sql

SELECT Juhtumi_nr, Name, MIN(SQRT(POW(Longitude - Lon, 2) + POW(Latitude - Lat, 2)) * 111.3) AS Kaugus
FROM liiklusonnetus, seadmed
GROUP BY Juhtumi_nr LIMIT 10;
```

```python
%%sql


Create VIEW v_koord_kaugused AS SELECT Juhtumi_nr, Name, MIN(SQRT(POW(Longitude - Lon, 2) + POW(Latitude - Lat, 2)) * 111.3) AS Kaugus
FROM liiklusonnetus, seadmed
GROUP BY Juhtumi_nr LIMIT 10;
```

```python
%%sql
SELECT* FROM v_koord_kaugused Limit 5;
```

```python
%%sql

UPDATE liiklusonnetus 
   SET closest_feature=(SELECT Name FROM v_koord_kaugused WHERE liiklusonnetus.Juhtumi_nr=v_koord_kaugused.Juhtumi_nr);
```

```python
%%sql
SELECT * FROM liiklusonnetus LIMIT 10;
```

```python
%%sql

UPDATE liiklusonnetus 
   SET distance=(SELECT Kaugus FROM v_koord_kaugused WHERE liiklusonnetus.Juhtumi_nr=v_koord_kaugused.Juhtumi_nr);
```
