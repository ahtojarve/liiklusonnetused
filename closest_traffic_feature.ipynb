```python
%%sql

UPDATE liiklusonnetus
SET closest_feature = (
    SELECT seadmed.Name
    FROM seadmed
    ORDER BY (
        (seadmed.Lat - liiklusonnetus.Latitude) * (seadmed.Lat - liiklusonnetus.Latitude) +
        (seadmed.Lon - liiklusonnetus.Longitude) * (seadmed.Lon - liiklusonnetus.Longitude)
    ) ASC
    LIMIT 1
)
WHERE EXISTS (
    SELECT 1
    FROM seadmed
    WHERE (
        (seadmed.Lat - liiklusonnetus.Latitude) * (seadmed.Lat - liiklusonnetus.Latitude) +
        (seadmed.Lon - liiklusonnetus.Longitude) * (seadmed.Lon - liiklusonnetus.Longitude)
    ) = (
        SELECT MIN(
            (seadmed.Lat - liiklusonnetus.Latitude) * (seadmed.Lat - liiklusonnetus.Latitude) +
            (seadmed.Lon - liiklusonnetus.Longitude) * (seadmed.Lon - liiklusonnetus.Longitude)
        )
        FROM seadmed
    )
);
```
